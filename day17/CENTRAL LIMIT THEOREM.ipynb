```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Create a population (right-skewed)
population = np.random.exponential(scale=10, size=10000)

# Plot population distribution
plt.hist(population, bins=20)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Population Distribution (Non-Normal)")
plt.show()
```

```python
sample_size = 30        # size of each sample
num_samples = 1000      # number of samples

sample_means = []

for i in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)
```

```python
plt.hist(sample_means, bins=10)
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Distribution of Sample Means (CLT)")
plt.show()
```

```python
plt.figure(figsize=(12,5))

# Population
plt.subplot(1, 2, 1)
plt.hist(population, bins=50)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Population Distribution")

# Sample Means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=40)
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Sample Mean Distribution (CLT)")

plt.tight_layout()
plt.show()
```
