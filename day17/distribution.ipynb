```python
import numpy as np
import matplotlib.pyplot as plt 

data = np.array([10, 12, 13, 14, 15, 16, 17, 18, 19, 100, 105])
```

```python
data
```

```python
# to calculate the mean 
mean = np.mean(data)
#to calculate the standard deviation
std_dev = np.std(data)
```

```python
plt.figure()
plt.boxplot(data)
plt.show()
```

```python
z_scores = (data - mean) / std_dev
print("Z-scores:", z_scores)
print("Mean:", mean)
print("Standard Deviation:", std_dev)
```

```python
plt.figure()
plt.hist(data)
plt.title("Histogram of Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Given population (skewed)
population = np.array([2, 3, 1, 4, 2, 6, 1, 9, 12, 20, 25, 30])

def sample_means(pop, n, simulations=1000):
    means = []
    for _ in range(simulations):
        sample = np.random.choice(pop, size=n, replace=True)
        means.append(np.mean(sample))
    return means

# n = 5
means_n5 = sample_means(population, 5)

plt.figure()
plt.hist(means_n5, bins=30)
plt.title("Sampling Distribution of Mean (n=5)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

# n = 30
means_n30 = sample_means(population, 30)

plt.figure()
plt.hist(means_n30, bins=30)
plt.title("Sampling Distribution of Mean (n=30)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create uniform population
uniform_population = np.random.uniform(0, 100, 100000)

# n = 2
means_n2 = sample_means(uniform_population, 2)

plt.figure()
plt.hist(means_n2, bins=30)
plt.title("Uniform Population - Sampling Distribution (n=2)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

# n = 10
means_n10 = sample_means(uniform_population, 10)

plt.figure()
plt.hist(means_n10, bins=30)
plt.title("Uniform Population - Sampling Distribution (n=10)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

# n = 50
means_n50 = sample_means(uniform_population, 50)

plt.figure()
plt.hist(means_n50, bins=30)
plt.title("Uniform Population - Sampling Distribution (n=50)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
sigma = 20

sample_sizes = [4, 16, 64]

for n in sample_sizes:
    standard_error = sigma / np.sqrt(n)
    print(f"Sample Size = {n}, Standard Error = {standard_error}")
```

```python
mean = 50
sigma = 4
n = 36

standard_error = sigma / np.sqrt(n)

z = 1.96  # 95% confidence

margin = z * standard_error

lower = mean - margin
upper = mean + margin

print("Standard Error:", standard_error)
print("95% Confidence Interval:", (lower, upper))
```

```python
# LLN demonstration
true_mean = 100
population_normal = np.random.normal(true_mean, 20, 100000)

sample_sizes = [5, 20, 100, 500]
means_progress = []

for n in sample_sizes:
    means = sample_means(population_normal, n)
    means_progress.append(np.mean(means))

plt.figure()
plt.plot(sample_sizes, means_progress)
plt.title("Law of Large Numbers Demonstration")
plt.xlabel("Sample Size")
plt.ylabel("Average of Sample Means")
plt.show()


# CLT distribution for large n
means_n100 = sample_means(population_normal, 100)

plt.figure()
plt.hist(means_n100, bins=30)
plt.title("CLT Demonstration - Sampling Distribution (n=100)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
