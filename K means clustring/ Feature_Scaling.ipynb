```python
import pandas as pd
import kagglehub
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Download dataset
path = kagglehub.dataset_download("arjunbhasin2013/ccdata")

# Load dataset
df = pd.read_csv(path + "/CC GENERAL.csv")

# Select required features
X = df[["PURCHASES", "PURCHASES_FREQUENCY"]]

# Handle missing values
X = X.ffill()

# ---------------------------------
# Step A: K-Means WITHOUT Scaling
# ---------------------------------
kmeans1 = KMeans(n_clusters=3, random_state=42)
y_kmeans1 = kmeans1.fit_predict(X)

plt.figure()
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_kmeans1)
plt.title("K-Means WITHOUT Scaling")
plt.xlabel("PURCHASES")
plt.ylabel("PURCHASES_FREQUENCY")
plt.show()

# ---------------------------------
# Step B: K-Means WITH Scaling
# ---------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans2 = KMeans(n_clusters=3, random_state=42)
y_kmeans2 = kmeans2.fit_predict(X_scaled)

plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans2)
plt.title("K-Means WITH Scaling")
plt.xlabel("Scaled PURCHASES")
plt.ylabel("Scaled PURCHASES_FREQUENCY")
plt.show()
```
