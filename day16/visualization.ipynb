```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# 1. Normal Distribution (Heights)
heights = pd.Series(np.random.normal(loc=170, scale=10, size=1000))

plt.figure()
plt.hist(heights, bins=30, density=True)
heights.plot(kind='kde')
plt.title("Normal Distribution - Heights")
plt.show()

print("Normal Distribution")
print("Mean:", heights.mean())
print("Median:", heights.median())

# 2. Right-Skewed Distribution (Incomes)
incomes = pd.Series(np.random.exponential(scale=50000, size=1000))

plt.figure()
plt.hist(incomes, bins=30, density=True)
incomes.plot(kind='kde')
plt.title("Right-Skewed Distribution - Incomes")
plt.show()

print("\nRight-Skewed Distribution")
print("Mean:", incomes.mean())
print("Median:", incomes.median())

# 3. Left-Skewed Distribution (Exam Scores)
scores = 100 - np.random.exponential(scale=10, size=1000)
scores = np.clip(scores, 0, 100)
scores = pd.Series(scores)

plt.figure()
plt.hist(scores, bins=30, density=True)
scores.plot(kind='kde')
plt.title("Left-Skewed Distribution - Exam Scores")
plt.show()

print("\nLeft-Skewed Distribution")
print("Mean:", scores.mean())
print("Median:", scores.median())
```
