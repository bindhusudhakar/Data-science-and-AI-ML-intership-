```python
import pandas as pd
import kagglehub
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Download dataset
path = kagglehub.dataset_download("iabhishekofficial/mobile-price-classification")
print("Dataset path:", path)

import os
print(os.listdir(path))

# Load data
df = pd.read_csv(f"{path}/train.csv")
print(df.head())

# Preprocess
X = df.drop('price_range', axis=1)
y = df['price_range']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Test different C values
for C in [0.01, 0.1, 1, 10, 100]:
    svm = SVC(kernel='rbf', C=C)
    svm.fit(X_train, y_train)
    print(f"C={C}, Accuracy={accuracy_score(y_test, svm.predict(X_test)):.4f}")

# GridSearchCV for best params
params = {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['rbf', 'linear']}
grid = GridSearchCV(SVC(), params, cv=5)
grid.fit(X_train, y_train)
print("Best Params:", grid.best_params_)
print("Best Accuracy:", grid.best_score_)
```
