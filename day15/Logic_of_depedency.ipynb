```python
import random

trials = 10000
count_success = 0

for _ in range(trials):
   
    coin = random.choice(["H", "T"])
  
    dice = random.randint(1, 6)
    
    if coin == "H" and dice == 6:
        count_success += 1

experimental_probability = count_success / trials

print("Experimental Probability (Heads AND 6):", experimental_probability)
```

```python
import random

trials = 10000
count_both_red = 0

for _ in range(trials):

    bag = ["R"] * 5 + ["B"] * 5
   
    first = random.choice(bag)
    bag.remove(first)   
   
    second = random.choice(bag)
    
    if first == "R" and second == "R":
        count_both_red += 1

experimental_probability = count_both_red / trials

print("Experimental Probability (Both Red):", experimental_probability)
```
